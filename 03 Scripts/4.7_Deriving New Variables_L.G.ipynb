```python
# Define a function using If statements

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

### Defining busiest periods of time labeled 'Most orders = 10, 11, 14, 15, 13, 12, 16, 9';  'Average orders = 17, 8, 18, 19, 20, 7, 21, 22';  and 'Fewest orders = 23, 6, 0,1, 5, 2, 4, 3'.

```python
# Create a new column containing these labels called 'order_hour_of_day'

order_hour_of_day = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    order_hour_of_day.append('Most orders')
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    order_hour_of_day.append('Fewest orders')
  else:
    order_hour_of_day.append('Average orders')
```

```python
# Add dataframe as 'busiest_period_of_day' column
df_ords_prods_merged['busiest_period_of_day'] = order_hour_of_day
```
