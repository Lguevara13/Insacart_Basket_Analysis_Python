```python
#Import data set

orders_products_include = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_cust_all.pkl'))
```

```python
orders_products_include.shape
```

```python
#Create a scatterplot that plots prices against prices to look for anomolies.

sctplt = sns.scatterplot(x = 'prices', y = 'prices', data = orders_products_include)

# Format x-axis tick labels as dollars
def format_dollars(value, tick_number):
    return f'${int(value):,}' 

# Format y-axis tick labels as dollars
def format_dollars(value, tick_number):
    return f'${int(value):,}'  # Format as dollars with thousands separators

sctplt.yaxis.set_major_formatter((format_dollars))
sctplt.xaxis.set_major_formatter((format_dollars))

plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Saving the scatterplot

sctplt.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'sctplt_prices_include.png'))
```

```python
#Create scatterplot of the entire merged data frame plotting age and income.

sctrplt2 = sns.scatterplot(x= 'age', y = 'income', data = orders_products_include)

#Set custom axis labels
sctrplt2.set_xlabel("Age")  
sctrplt2.set_ylabel("Income")

# Format y-axis tick labels as dollars
def format_dollars(value, tick_number):
    return f'${int(value):,}'  # Format as dollars with thousands separators

sctrplt2.yaxis.set_major_formatter((format_dollars))

plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Save scatterplot (sctrplt2)

sctrplt2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'sctrplt_age_income_include.png'))
```
